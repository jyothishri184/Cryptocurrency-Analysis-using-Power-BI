```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
df_train = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
df_test = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
```

```python
df_train
```

```python
df_test
```

```python
passenger_id = df_test['PassengerId']
```

```python
final_df = df_train.drop(['PassengerId','Cabin','Name'],axis=1)

final_df
```

```python
final_df = final_df.dropna()
final_df
```

```python
final_df["Sex"].replace(final_df['Sex'].unique(),value=[1,0],inplace=True)
```

```python
final_df
```

```python
final_df = pd.get_dummies(final_df,drop_first=True)
final_df
```

```python
train_y = final_df["Survived"]
train_x = final_df[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare']]
```

```python
from sklearn.model_selection import train_test_split
train_data,val_data,train_target,val_target =  train_test_split(train_x,train_y,train_size=0.8)
```

```python
model=GaussianNB()
model.fit(train_data,train_target)
val_pred=model.predict(val_data)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(val_target,val_pred)*100)
```
